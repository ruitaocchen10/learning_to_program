```python
# rate limiter

class RateLimiter:
    def __init__(self, max_requests):
        self.max_requests = max_requests
        self.request_counts = {}

    def create_account(self):
        user = input("Enter your name:")
        if user in self.request_counts:
            print("User already exists")
            return
        else:
            self.request_counts[user] = 0

    def allow_requests(self, user_id):
        self.request_count = self.request_counts.get(user_id, 0)    
        if self.request_count < self.max_requests:
            return True
        else:
            return False

    def add_request(self, user_id):
        self.request_counts[user_id] = self.request_counts.get(user_id, 0) + 1

limiter = RateLimiter(max_requests=3)                                                 
                                                                                        
  # Test 1: basic blocking                                                              
print("-- Test 1: alice makes 5 requests (limit=3) --")                               
for i in range(5):                                                                    
    allowed = limiter.allow_requests("alice")                                         
    if allowed:                                                                       
        limiter.add_request("alice")                                                  
    print(f"Request {i+1}: {'ALLOWED' if allowed else 'BLOCKED'}")                    
                                                                                        
  # Test 2: separate users don't affect each other                                      
print("\n-- Test 2: bob is a separate user --")                                       
allowed = limiter.allow_requests("bob")                                               
if allowed:                                                                           
    limiter.add_request("bob")                                                        
print(f"Bob request 1: {'ALLOWED' if allowed else 'BLOCKED'}")                        
                                                                                        
  # Test 3: duplicate account                                                           
print("\n-- Test 3: duplicate account --")                                            
limiter.request_counts["carol"] = 0                                                   
limiter.create_account()
```
