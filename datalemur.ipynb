```python
def total_letters(N):
    nums = list(range(1,(N + 1)))
    sum = 0

    for item in nums:
        item = str(item)
        return item #have to convert into strings
        sum += len(item)

    return sum

N = 5
total_letters(N)
```

```python
def compound_interest(principal, rate, contribution, years):
    total = principal

    for year in range(years):
        total += (rate/100) * principal
        total += contribution
        principal = total
      
    return round(total, 2)

compound_interest(500, 10, 50, 3)
```

```python
def fizz_buzz_sum(target):
    factors = []
    nums = list(range(1, target))

    for i in nums:
        if i % 3 == 0:
            factors.append(i)
        elif i % 5 == 0:
            factors.append(i)

    sum = 0
    for i in factors:
        sum += i

    return sum

fizz_buzz_sum(10)
```

```python
def isPalindrome(phrase):
  phrase = phrase.lower()
  alphanumeric = ""
  for char in phrase:
    if char.isalnum():
        alphanumeric += char
  left = 0 
  right = len(alphanumeric) - 1
  
  while left < right:
    if alphanumeric[left] == alphanumeric[right]:
      left += 1
      right -= 1
    else:
      return False
      
  return True
```
