```python
# Given an integer array nums, return true if any value appears more than once in the array, otherwise return false.

class Solution:
    def hasDuplicate(self, nums):
        values = []
        for i in nums:
            if i not in values:
                values.append(i)
            else:
                return True
        
        return False


nums = [1,5,4,3,6]
solution = Solution()
answer = solution.hasDuplicate(nums)
print(answer)
```

```python
""" Given an array of integers nums, return the length of the longest consecutive sequence of elements that can be formed. """

class Solution:
    def longestConsecutive(self, nums):
        starters = []

        for n in nums:
            if n-1 not in nums:
                starters.append(n)

        count = 0

        for n in starters:
            nxt_val = n
            temp_count = 0
            while nxt_val in nums:
                nxt_val += 1
                if temp_count >= count:
                    count += 1
                    temp_count += 1
                else:
                    temp_count += 1

        return count
    
nums = [2,20,4,10,3,4,5]
solution = Solution()
answer = solution.longestConsecutive(nums) 
print(answer)
```
