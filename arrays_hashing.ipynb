```python
# Given an integer array nums, return true if any value appears more than once in the array, otherwise return false.

class Solution:
    def hasDuplicate(self, nums):
        values = []
        for i in nums:
            if i not in values:
                values.append(i)
            else:
                return True
        
        return False


nums = [1,5,4,3,6]
solution = Solution()
answer = solution.hasDuplicate(nums)
print(answer)
```
