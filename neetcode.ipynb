```python
class Solution:
    def hasDuplicate(self, nums):
        values = []
        for i in nums:
            if i in values:
                return True
            else:
                values.append(i)

        return False
    
nums = [1, 2, 3, 4, 5, 6, 6]
Solution.hasDuplicate(nums, nums)
```

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        s = sorted(s)
        t = sorted(t)
        p1 = 0
        p2 = 0

        if len(s) != len(t):
            return False

        while p1 < len(s) and p2 < len(t):
            if s[p1] == t[p2]:
                p1 += 1
                p2 += 1
            else: 
                return False
        
        return True

s = "racecar"
t = "carrace"
solution = Solution()
solution.isAnagram(s, t)
```

```python
class Solution:
    def twoSum(self, nums: list[int], target: int) -> list[int]:
        slow = 0
        
        while slow < len(nums) - 1:
            for i, fast in enumerate(nums):
                if slow != i and nums[slow] + fast == target:
                    return [slow, i]
            slow += 1

nums = [3,4,5,6]
target = 7
solution = Solution()
solution.twoSum(nums, target)
```

```python
class Solution:
    def groupAnagrams(self, strs: list[str]) -> list[list[str]]:
        result = {} # hashmap
        
        for s in strs:
            count = [0] * 26

            for c in s:
                count[ord[c] - ord["a"]] += 1 #ascii values
                
            result[count].append(s)

        return result.values()

strs = ["act","pots","tops","cat","stop","hat"]
solution = Solution()
solution.groupAnagrams(strs)
```

```python
class Solution:
    def hasDuplicate(self, nums):
        hashset = set()

        for n in nums:
            if n in hashset:
                return True
            hashset.add(n)

        return False
    
nums = [1, 2, 3, 4, 5, 6, 6]
Solution.hasDuplicate(nums, nums)
```
