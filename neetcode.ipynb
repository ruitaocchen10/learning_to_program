```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        s = sorted(s)
        t = sorted(t)
        p1 = 0
        p2 = 0

        if len(s) != len(t):
            return False

        while p1 < len(s) and p2 < len(t):
            if s[p1] == t[p2]:
                p1 += 1
                p2 += 1
            else: 
                return False
        
        return True

s = "racecar"
t = "carrace"
solution = Solution()
solution.isAnagram(s, t)
```

```python
class Solution:
    def twoSum(self, nums: list[int], target: int) -> list[int]:
        slow = 0
        
        while slow < len(nums) - 1:
            for i, fast in enumerate(nums):
                if slow != i and nums[slow] + fast == target:
                    return [slow, i]
            slow += 1

nums = [3,4,5,6]
target = 7
solution = Solution()
solution.twoSum(nums, target)
```

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        cleaned_string = ''.join(char.lower() for char in s if char.isalnum())
        
        l = 0
        r = len(cleaned_string) - 1

        while l < r:
            if cleaned_string[l] == cleaned_string[r]:
                l += 1
                r -= 1
            else:
                return False
            
        return True

s = Solution()
answer = s.isPalindrome("Was it a car or a cat I saw?")

print(answer)
```

```python
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        pairings = { "}" : "{", "]" : "[",  ")" : "(" }

        for item in s:
            if item in pairings:
                if stack and stack[-1] == pairings[item]:
                    stack.pop()

                else:
                    return False
        
            else:
                stack.append(item)

        return True if not stack else False

string = "{[]}"
solution = Solution()
answer = solution.isValid(string)

print(answer)
```

```python
class Solution:
    def twoSum(self, numbers, target):
        left = 0
        right = len(numbers) - 1

        sum = 0
        output = []
        while left < right:
            sum = numbers[left] + numbers[right]
            
            if sum > target: 
                right -= 1
            elif sum < target:
                left += 1
            else:
                output.append(left + 1)
                output.append(right + 1)
                return output

        
numbers = [1,2,3,4]
target = 3
solution = Solution()
answer = solution.twoSum(numbers, target)
print(answer)
```

```python
class Solution:
    def evalRPN(self, tokens):
        stack = []

        for char in tokens:
            if char not in {"+", "-", "*", "/"}:
                stack.append(int(char))
            else:
                b = stack.pop()
                a = stack.pop()

                if char == "+":
                    stack.append(a + b)
                elif char == "-":
                    stack.append(a - b)
                elif char == "*":
                    stack.append(a * b)
                else:
                    stack.append(int(a / b))  # truncate toward 0

        return stack[0]
    
```

```python
class Solution:
    def search(self, nums, target):
        l = 0
        r = len(nums) - 1
        
        while l <= r:
            m = (l + r) // 2

            if nums[m] > target:
                r = m - 1
            elif nums[m] < target:
                l = m + 1
            else:
                return m

        return -1
        
nums = [-1,0,2,4,6,8]
target = 4
solution = Solution()
answer = solution.search(nums, target)
print(answer)
```
