```python
""" Design a stack class that supports the push, pop, top, and getMin operations.

MinStack() initializes the stack object.
void push(int val) pushes the element val onto the stack.
void pop() removes the element on the top of the stack.
int top() gets the top element of the stack.
int getMin() retrieves the minimum element in the stack.
 """

class MinStack:
    def __init__(self):
        self.stack = []
        self.min_val = []

    def push(self, val: int) -> None:
        self.stack.append(val)
        if not self.min_val:
            self.min_val.append(val)
        elif val <= self.min_val[-1]:
            self.min_val.append(val)

    def pop(self) -> None:
        popped = self.stack.pop()

        if popped == self.min_val[-1]:
            self.min_val.pop()
        
    def top(self) -> int:
        return self.stack[-1]

    def getMin(self) -> int:
        return self.min_val[-1]

operations = ["MinStack", "push", "push", "push", "getMin", "pop", "top", "getMin"]
values = [None, 1, 2, 0, None, None, None, None]

solution = MinStack()

for op, val in zip(operations, values):
    if op == "push":
        solution.push(val)
    elif op == "pop":
        solution.pop()
    elif op == "top":
        print(solution.top())
    elif op == "getMin":
        print(solution.getMin())
```

```python
""" You are given an array of integers temperatures where temperatures[i] represents the daily temperatures on the ith day. """

class Solution:
    def dailyTemperatures(self, temperatures):
        result = []

        for i, temp in enumerate(temperatures):
            found = False
            for k in range(i + 1, len(temperatures)):
                if temperatures[k] > temp:
                    result.append(k-i) 
                    found = True
                    break
            if not found:
                result.append(0)

        return result


temperatures = [30,38,30,36,35,40,28]
solution = Solution()
answer = solution.dailyTemperatures(temperatures)
print(answer)
        
```

```python
""" You are given an array of integers temperatures where temperatures[i] represents the daily temperatures on the ith day. """

# Better solution

class Solution:
    def dailyTemperatures(self, temperatures):
        stack = []
        result = [0] * len(temperatures)

        for i, temp in enumerate(temperatures):
            while stack and temp > stack[-1][0]:
                StackTemp, StackInd = stack.pop()
                result[StackInd] = i - StackInd
            stack.append((temp, i))

        return result
        
temperatures = [30,30,38,36,35,40,28]
solution = Solution()
answer = solution.dailyTemperatures(temperatures)
print(answer)
        
```
