```python
# Class Practice

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.amount = amount
        self.balance += self.amount

    def withdraw(self, amount):
        self.amount = amount
        self.balance -= self.amount

    def get_balance(self):
        return self.balance


account = BankAccount("Ruitao", 0)

account.deposit(100)
account.withdraw(40)

print(account.get_balance())
```
