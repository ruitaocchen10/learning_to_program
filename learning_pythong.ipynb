```python
# min teams problem

def countTeams(teamSize_1, teamSize_2, p):
    # Write your code here
    l_team = max(teamSize_1, teamSize_2)
    s_team = min(teamSize_1, teamSize_2)
    
    max_large = p // l_team
    
    for l_count in range(max_large, -1, -1):
        remainder = p - (l_count * l_team)
        
        if remainder % s_team == 0:
            s_count = remainder / s_team
            total_teams = l_count + s_count
            return total_teams
            
    return -1
```

```python
# get health records

import requests

def healthCheckup(lowerlimit, upperlimit):
    base_url = "https://jsonmock.hackerrank.com/api/medical_records"   
    
    all_records = []

    response = requests.get(f"{base_url}?page=1")
    first_result = response.json()

    total_pages = first_result["total_pages"]

    page = 1

    while page < total_pages + 1:
        response = requests.get(f"{base_url}?page={page}")
        result = response.json()

        all_records.extend(result["data"])

        page += 1

    count_in_range = 0

    for item in all_records:
        if lowerlimit < item["vitals"]["bloodPressureDiastole"] < upperlimit:
            count_in_range += 1

    return count_in_range

healthCheckup(120, 160)
```
