```python
#Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

#You may assume that each input would have exactly one solution, and you may not use the same element twice.

#You can return the answer in any order.

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        hashmap = {} # Empty dictionary to store each number and index
        for i in range(len(nums)): #range stores nums as indices
            hashmap[nums[i]] = i #Stores key nums[i] (which is the number in the array), and value i (index)
        for i in range(len(nums)): #For each value in nums
            complement = target - nums[i] #Creates a compliment value for each num
            if complement in hashmap and hashmap[complement] != i: #if compliment is in the hashmap and compliment is not = existing number
                return [i, hashmap[complement]] #return the list of the two indices
        # If no valid pair is found, return an empty list
        return []
```

```python
string = "A man, a plan, a canal: Panama"

def palindrome(string):
    #string_split = string.split() 
    string = string.lower() #convert to lowercase
    
    front = 0
    back = len(string) - 1

    while front < back:
        if not string[front].isalnum(): #skips non-alphanumeric char from front
            front += 1
            continue

        if not string[back].isalnum(): #skips non-alphanumeric char from back
            front += 1
            continue

        if string[front] != string[back]:
            return False

        front += 1
        back -= 1

    return True

palindrome(string)
```

```python
def form(numbers, target):
    start = 0
    end = len(numbers) - 1

    while start < end: #while loop 
        current_sum = numbers[start] + numbers[end]

        if current_sum == target:
            return [start + 1, end + 1]
        elif current_sum > target:
            end -= 1
        else:
            start += 1  

    return []              
        
numbers = [2, 7, 11, 15]
target = 9

form(numbers, target)
```

```python
# reversing a string array 

def reverse(s):
    start = 0
    end = len(s) - 1

    while start < end:
        s[start], s[end] = s[end], s[start]
        # temp = s[start]
        # s[start] = s[end]
        # s[end] = temp
        start += 1
        end -= 1

    return s

s = ["h", "e", "l", "l", "o"]

print(reverse(s))

```

```python
def swap(nums):
    slow = 0

    for fast in range(1, len(nums)):
        if nums[fast] != 0:
            nums[slow], nums[fast] = nums[fast], nums[slow]
            slow += 1

    return nums

nums = [0, 1, 0, 3, 12]
swap(nums)
```

```python
def squares(nums):
    n = len(nums)
    result = [0] * n
    start = 0
    end = n - 1
    position = n - 1

    while start <= end:
        start_squared = nums[start] ** 2
        end_squared = nums[end] ** 2
        
        if start_squared > end_squared:
            result[position] = start_squared
            start += 1
        else:
            result[position] = end_squared
            end -= 1
        
        position -= 1

    return result

nums = [-4, -1, 0, 3, 10]
print(squares(nums))
```

```python
def area(height):
    left = 0
    right = len(height) - 1
    area = 0

    while left < right:
        w = right - left
        h = min(height[left], height[right])
        a = w * h
        if a > area:
            area = a
            left += 1
        else:
            right -=1

    return area

height = [1, 8, 6, 2, 5, 4, 8, 3, 7]
area(height)

```

```python
def match(nums1, nums2):
    nums1.sort()
    nums2.sort()
    first = 0
    second = 0
    output = []

    while first < len(nums1) and second < len(nums2):
        if nums1[first] == nums2[second]:
            output.append(nums1[first])
            first += 1
            second += 1
        elif nums1[first] > nums2[second]:
            second += 1
        else:
            first += 1

    return output

nums1 = [4, 9, 5]
nums2 = [9, 4, 9, 8, 4]
match(nums1, nums2)
```

```python
def find(nums):
    nums.sort()
    n = len(nums)
    r = range(1, n + 1)
    orlis = list(r)
    output = []
    nu = 0
    lu = 0

    while lu < len(orlis):
        if nu < len(nums) and nums[nu] == orlis[lu]:
            nu += 1
            lu += 1
        elif nu < len(nums) and nums[nu] < orlis[lu]:
            nu += 1
        elif nums[nu] > orlis[lu]:
            output.append(orlis[lu])
            lu += 1
    
    return output


nums = [4, 3, 2, 7, 8, 2, 3, 1]
find(nums)
```

```python
def backspace(s,t):
    first = 0
    second = 0
    a_result = []
    b_result = []

    while first < len(s) and second < len(t):
        a_result.append(s[first])
        b_result.append(t[second])
        first += 1
        second += 1
        if s[first] == '#':
            a_result.pop(s[first-1])
        elif t[second] == '#':
            b_result.pop(t[second-1])

    if a_result == b_result:
        return True
    else:
        return False
```

```python
def shortest(s, c):
    output = []

    for i in range(len(s)):
        if s[i] == c:
            output.append(0)
            continue

        distance = 0

        right = i + 1
        left = i - 1

        while left >= 0 and right < len(s):
            if left >= 0 and s[left] == c:
                distance = abs(left)
                break
            if right < len(s) and s[right] == c:
                distance = right
                break
            right += 1
            left -= 1

        output.append(distance)

    return output

s = "loveleetcode"
c = "e"

shortest(s, c)
            
# still haven't gotten this question
```

```python
def r_vowels(s):
    s = list(s)
    left = 0
    right = len(s) - 1
    vowels = ('a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U')


    while left < right:
        if s[left] not in vowels:
            left += 1
        elif s[left] in vowels:
            if s[right] not in vowels:
                right -= 1
            elif s[right] in vowels:
                s[left], s[right] = s[right], s[left]
                right -= 1
                left += 1
    s = ''.join(s)

    return s

s = "hello"
r_vowels(s)
```
