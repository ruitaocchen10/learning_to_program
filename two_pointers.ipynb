```python
#Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

#You may assume that each input would have exactly one solution, and you may not use the same element twice.

#You can return the answer in any order.

class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        hashmap = {} # Empty dictionary to store each number and index
        for i in range(len(nums)): #range stores nums as indices
            hashmap[nums[i]] = i #Stores key nums[i] (which is the number in the array), and value i (index)
        for i in range(len(nums)): #For each value in nums
            complement = target - nums[i] #Creates a compliment value for each num
            if complement in hashmap and hashmap[complement] != i: #if compliment is in the hashmap and compliment is not = existing number
                return [i, hashmap[complement]] #return the list of the two indices
        # If no valid pair is found, return an empty list
        return []
```

```python
string = "A man, a plan, a canal: Panama"

def palindrome(string):
    #string_split = string.split() 
    string = string.lower() #convert to lowercase
    
    front = 0
    back = len(string) - 1

    while front < back:
        if not string[front].isalnum(): #skips non-alphanumeric char from front
            front += 1
            continue

        if not string[back].isalnum(): #skips non-alphanumeric char from back
            front += 1
            continue

        if string[front] != string[back]:
            return False

        front += 1
        back -= 1

    return True

palindrome(string)
```

```python
def form(numbers, target):
    start = 0
    end = len(numbers) - 1

    while start < end: #while loop 
        current_sum = numbers[start] + numbers[end]

        if current_sum == target:
            return [start + 1, end + 1]
        elif current_sum > target:
            end -= 1
        else:
            start += 1  

    return []              
        
numbers = [2, 7, 11, 15]
target = 9

form(numbers, target)
```

```python
# reversing a string array 

def reverse(s):
    start = 0
    end = len(s) - 1

    while start < end:
        s[start], s[end] = s[end], s[start]
        # temp = s[start]
        # s[start] = s[end]
        # s[end] = temp
        start += 1
        end -= 1

    return s

s = ["h", "e", "l", "l", "o"]

print(reverse(s))

```

```python
def swap(nums):
    slow = 0

    for fast in range(1, len(nums)):
        if nums[fast] != 0:
            nums[slow], nums[fast] = nums[fast], nums[slow]
            slow += 1

    return nums

nums = [0, 1, 0, 3, 12]
swap(nums)
```

```python
def squares(nums):
    n = len(nums)
    result = [0] * n
    start = 0
    end = n - 1
    position = n - 1

    while start <= end:
        start_squared = nums[start] ** 2
        end_squared = nums[end] ** 2
        
        if start_squared > end_squared:
            result[position] = start_squared
            start += 1
        else:
            result[position] = end_squared
            end -= 1
        
        position -= 1

    return result

nums = [-4, -1, 0, 3, 10]
print(squares(nums))
```

```python
def area(height):
    left = 0
    right = len(height) - 1
    area = 0

    while left < right:
        width = right - left
        height = min(height[left], height[right])
        a = width * height
        if a < area:
            left += 1
        elif a > area:
            area = a
            right -=1

    return area

height = [1, 8, 6, 2, 5, 4, 8, 3, 7]
area(height)

```
